Over the next few questions, we're going to make a bar blot of a subset of the regions in Canada, the Atlantic and the Prairies. All other regions will be considered collectively as "Other".

The following code makes a list out of the `'region'` column and then replaces every non-Atlantic non-Prairies region with `'Other'`.

That's an if statement! And a fancy version of a for loop that lets us use `i` as an index into the `region` list.

```python
region = list(prov_data["region"])

for i in range(len(region)):
    if region[i] != "Atlantic" and region[i] != "Prairies":
        region[i] = "Other"
        
region
```
